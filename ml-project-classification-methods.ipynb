```python
data = pd.read_csv('/kaggle/input/voicegender/voice.csv')
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.describe()
```

```python
print(data.isnull().sum())
```

```python
x = data.drop('label', axis = 1)
```

```python
x.corr()
```

```python
data.label = [1 if each == "male" else 0 for each in data.label]
```

```python
data.corr()
```

```python
y = data.label.values
x_data = data.drop(["label"], axis = 1)
```

```python
x = (x_data - np.min(x_data)) / (np.max(x_data))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
#test_size =0.2 means %20 test datas ,80% train datas
method_names = []
method_scores = []
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
print("Logistic Regression Classification Test Accuracy : {}".format(log_reg.score(x_test, y_test )))
method_names.append("Logistic Reg")
method_scores.append(log_reg.score(x_test, y_test ))

#conf matrix
y_pred = log_reg.predict(x_test)
conf_matrix = confusion_matrix( y_test,y_pred )

f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_matrix, annot = True, linewidths = 0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3,  metric = 'manhattan')
knn.fit(x_train, y_train)
print("Score for number of neighbors = 3: {}".format(knn.score(x_test, y_test )))
method_names.append("KNN")
method_scores.append(knn.score(x_test, y_test ))

#conf matrix
y_pred = knn.predict(x_test)
conf_matrix = confusion_matrix( y_test,y_pred )

f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_matrix, annot = True, linewidths = 0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
grid = { 
    'learning_rate':[0.2,0.4,0.6],
    'n_estimators' :np.arange(300,500,999)
}
gb = GradientBoostingClassifier()
gb_cv = GridSearchCV(gb, grid, cv = 4)
gb_cv.fit(x_train, y_train)
print("Best Parameters :",gb_cv.best_params_)
print("Train Score :",gb_cv.best_score_)
print("Test Score :",gb_cv.score(x_test, y_test))
print("Gradient Forest CV Classification Score is: {}".format(gb_cv.score(x_test, y_test )))
method_names.append(" Gradient Forest CV")
method_scores.append(gb_cv.score(x_test, y_test ))

y_pred = gb_cv.predict(x_test)
conf_matrix = confusion_matrix( y_test,y_pred )

f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_matrix, annot = True, linewidths = 0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()
```
